```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
import re
```

```python
links = []
titles = []
dates = []

for i in range(26968):
    links_temp = []
    titles_temp = []
    dates_temp = []
    
    url = 'https://www.hankyung.com/economy?page='+str(i+1)
    res = requests.get(url)
    html_soup = BeautifulSoup(res.text)
    a_list = html_soup.select('ul[class=list_basic] h3 a[href]')

    links_temp = [a['href'] for a in a_list]
    titles_temp = [a.text for a in a_list]
    dates_temp = [date.text for date in html_soup.select('ul[class=list_basic] span[class=time]')]
    
    links += links_temp
    titles += titles_temp
    dates += dates_temp
    
    if i % 100 == 0:
        print(titles_temp[0])
        if i % 5000 == 0:
            df = pd.DataFrame(columns=['title','date','link'])
            df['title'] = titles
            df['link'] = links
            df['date'] = dates
            df.to_csv('hankyung_links.csv', encoding='utf-8', index=False)
    time.sleep(np.random.randint(20,150)/101)

df = pd.DataFrame(columns=['title','date','link'])
df['title'] = titles
df['link'] = links
df['date'] = dates
df.to_csv('hankyung_econ_links.csv', encoding='utf-8', index=False)
```

```python
# df = pd.read_csv('hankyung_econ_links.csv', encoding='utf-8')
df = pd.read_csv('hankyung_econ_articles.csv', encoding='utf-8')
df['article'] = ''
df['paper'] = 'hankyung'
```

```python
err_list = []
for idx, row in df[df.article.isnull()].iterrows():
    url = row['link']
    res = requests.get(url)
    res.encoding='utf-8'
    html_soup = BeautifulSoup(res.text, 'lxml')
    if url.startswith('http://marketinsight'):
        a_list = html_soup.select('div[id=newsView]')
    else:
        a_list = html_soup.select('div[id=articletxt]')
    try:
        if len(a_list[0].text.split('/')) == 1:
            article = a_list[0].text
        else:
            article = ' '.join(a_list[0].text.split('/')[:-1])
            paper = re.sub('\s\s+', ' ', re.sub('\n|\t|\r', '', a_list[0].text.split('/')[-1]))
            df.loc[idx, 'paper'] = paper
    except:
        article = ''
        err_list += [idx]
    article = re.sub('\s\s+', ' ', re.sub('\n|\t|\r', '', article))
    df.loc[idx, 'article'] = article
    
    if idx % 100 == 0:
        print(row.title)
        if idx % 1000 == 0:
            df.to_csv('hankyung_econ_articles.csv', encoding='utf-8', index=False)
    time.sleep(np.random.randint(20,200)/101)
df.to_csv('hankyung_econ_articles.csv', encoding='utf-8', index=False)
```
