```python
from bs4 import BeautifulSoup
from io import BytesIO
import datetime
import pandas as pd
import numpy as np
import requests
import json
import time
import re
```

```python
url = 'http://marketdata.krx.co.kr/contents/COM/GenerateOTP.jspx'
data = {
    'name': 'fileDown',
    'filetype': 'csv',
    'url': 'MKD/13/1302/13020303/mkd13020303',
#     'isu_cdnm': 'A005930/',
    'isu_cd': 'KR7005930003',
#     'isu_nm': '',
#     'isu_srt_cd': 'A005930',
    'period_selector': 'day',
    'fromdate': '',
    'todate': '',
    'pagePath': '/contents/MKD/13/1302/13020303/MKD13020303.jsp'
}

url_download = 'http://file.krx.co.kr/download.jspx'
headers_download = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; \
    x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36',
    'Content-Type': 'application/x-www-form-urlencoded',
    'Referer': 'http://marketdata.krx.co.kr/mdi'
}

d1 = datetime.date(2009, 4, 17)
d2 = datetime.date(2020, 3, 1)
dates = [d1 + datetime.timedelta(days=x) for x in range((d2-d1).days + 1)]

for date in dates:
    date_string = date.strftime('%Y%m%d')
    data['fromdate'] = date_string
    data['todate'] = date_string
    print(date.strftime('%Y%m%d'))

    r = requests.get(url, data, headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; \
        x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36'})
    res = requests.post(url_download, {'code':r.content}, headers=headers_download)
    df_investor = pd.read_csv(BytesIO(res.content))
    df_investor['ISIN'] = 'KR7005930003'
    df_investor['date'] = date
    df_investor.to_excel('./KRX_investor_data/KR7005930003_'+date_string+'.xlsx', encoding='utf-8', index=False)
    time.sleep(np.random.randint(50,200)/101)
```
