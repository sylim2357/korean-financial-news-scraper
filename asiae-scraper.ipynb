```python
import requests
from bs4 import BeautifulSoup, Comment, NavigableString, Tag
import bs4
import pandas as pd
import numpy as np
import time
import re
```

```python
SECTION_MAX_LENGTH_DICT = {'finance': 4684, 'world-economy': 3065}
section = 'finance'
# def make_links_df(section):
#     assert section in SECTION_MAX_LENGTH_DICT.keys()
df = pd.DataFrame(columns=['title','section','date','link'])
links = []
titles = []
sections = []
dates = []

for i in range(SECTION_MAX_LENGTH_DICT[section]):
    links_temp = []
    titles_temp = []
    sections_temp = []
    journalists_temp = []
    dates_temp = []

    url = 'https://www.asiae.co.kr/list/' + section + '/' + str(i+1)
    res = requests.get(url)
    res.encoding = 'utf-8'
    html_soup = BeautifulSoup(res.text)
    a_list = [a for a in (div.find('a') for div in html_soup.findAll('h3', {'class': 'l_fsttit'})) if a.text]

    links_temp = [a['href'] for a in a_list]
    titles_temp = [a['title'] for a in a_list]
#         sections_temp = [section]
    dates_temp = [date.text for date in html_soup.findAll('span', {'class':'txt_time'})]

    links += links_temp
    titles += titles_temp
#         sections += sections_temp
    dates += dates_temp

    if i % 100 == 0:
        print(titles[-2:])
#         print(sections[-2:])
        print(links[-2:])
    time.sleep(np.random.randint(50,200)/101)

df['title'] = titles
#     df['section'] = sections
df['link'] = links
df['date'] = dates
    
#     return df
```

```python
df['section'] = 'finance'
df['link'] = 'https:' + df['link']
```

```python
df.to_csv('asiae_finance_links.csv', encoding='utf-8', index=False)
```

```python
for idx, link in df.iterrows():
    url = link['link']
    res_article = requests.get(url)
    res_article.encoding = 'utf-8'
    html_article_soup = BeautifulSoup(res_article.text, 'lxml')
    article_raw = html_article_soup.find_all('div', {'class':'article'})
    article_temp = [article.text.strip() for article in article_raw]    
    article_temp = article_temp[0].split('(function(d,a,b,l,e,_)')[0]
    article = re.sub('\s\s+', ' ', re.sub('\n|\t|\r', '', article_temp))
    df.loc[idx, 'article'] = article
    
    if idx % 100 == 0:
        print(df.loc[idx, 'date'])
    time.sleep(np.random.randint(50,150)/101)
```

```python
df.to_csv('asiae_finance_articles.csv', encoding='utf-8', index=False)
```

```python
for idx, link in df[df.article.isnull()].iterrows():
    url = link['link']
    res_article = requests.get(url)
    res_article.encoding = 'utf-8'
    html_article_soup = BeautifulSoup(res_article.text, 'lxml')
    article_raw = html_article_soup.find('div', {'id':'txt_area'})
    if article_raw is not None:
        start = None
        end = None
        text = []
        if article_raw is not None:
            for c, child in enumerate(article_raw):
                if isinstance(child, Comment):
                    if child.strip() == '기사 부제목 영역':
                        start = c
                    if child.strip() == '본문//':
                        end = c
                        break
                    if child.strip() == '기사 부제목 영역//':
                        continue
                    if child.strip() == '본문':
                        continue
                if (start is not None) & (end is None):
                    if start < c:
                        if isinstance(child, Tag):
                            if '\xa0기자\xa0' not in child.text:
                                text.append(child.text.strip())
                        else:
                            if '\xa0기자\xa0' not in child:
                                text.append(child)
            article_txt = re.sub('\s\s+', ' ', re.sub('\n|\t|\r|\xa0', ' ', ' '.join(text))).strip()
            df.loc[idx, 'article'] = article_txt
    
    if idx % 100 == 0:
        print(df.loc[idx, 'date'])
    time.sleep(np.random.randint(50,100)/101)
```

```python
df.to_csv('asiae_finance_articles.csv', encoding='utf-8', index=False)
```
