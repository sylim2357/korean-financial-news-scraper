```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
import re
```

```python
df = pd.DataFrame(columns=['title','date','link'])
links = []
titles = []
dates = []

for i in range(12859):
    links_temp = []
    titles_temp = []
    dates_temp = []
    
    url = 'https://www.thebell.co.kr/free/content/article.asp?page='+str(i+1)
    res = requests.get(url)
    html_soup = BeautifulSoup(res.text)
    a_list = [a for a in (dl.find('a') for dl in html_soup.findAll("dl")) if a]
#     date_list = html_soup.findAll('span', {'class':'date'})

    links_temp = [a['href'] for a in a_list]
    titles_temp = [a['title'] for a in a_list]
    dates_temp = [date.text for date in html_soup.findAll('span', {'class':'date'})]
    
    links += links_temp
    titles += titles_temp
    dates += dates_temp
    
    if i % 100 == 0:
        print(titles_temp[0])
        if i % 1000 == 0:
            df.to_csv('the_bell_links.csv', encoding='utf-8', index=False)
    time.sleep(np.random.randint(30,150)/101)

df['title'] = titles
df['link'] = links
df['date'] = dates
```

```python
df['paper_name'] = 'the_bell'
df['link'] = ['http://www.thebell.co.kr/free/content/' + link[0] for link in df['link'].str.split('&').tolist()]
df['article'] = ''
```

```python
for idx, link in df.iterrows():
    url = link['link']
    res_article = requests.get(url)
    html_article_soup = BeautifulSoup(res_article.text)
    article = re.sub('\s\s+', ' ', re.sub('\n|\t|\r', '', html_article_soup.find('div',{'class':'viewSection'}).text))
    df.loc[idx, 'article'] = article
    
    if idx % 100 == 0:
        print(idx)
    time.sleep(np.random.randint(50,200)/101)
```
