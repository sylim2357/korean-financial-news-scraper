```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
import re
```

```python
df = pd.DataFrame(columns=['title','journalist','section','date','link'])
links = []
titles = []
sections = []
journalists = []
dates = []

for i in range(2731):
    links_temp = []
    titles_temp = []
    sections_temp = []
    journalists_temp = []
    dates_temp = []
    
    url = 'http://www.kbanker.co.kr/news/articleList.html?page='+str(i+1)+'&view_type=sm'
    res = requests.get(url)
    html_soup = BeautifulSoup(res.text, 'lxml')
    a_list = [a for a in (div.find('a') for div in html_soup.findAll('div', {'class':'list-titles'})) if a.text]
    dates_list = list(map(list, zip(*[date.text.split(' | ') for date in html_soup.findAll('div', {'class':'list-dated'}) if date.text])))
    if len(a_list) != len(dates_list[0]):
        res = requests.get(url)
        html_soup = BeautifulSoup(res.text, 'html5lib')
        a_list = [a for a in (div.find('a') for div in html_soup.findAll('div', {'class':'list-titles'})) if a.text]
#         dates_list = list(map(list, zip(*[date.text.split(' | ') for date in html_soup.findAll('div', {'class':'list-dated'}) if date.text])))
    assert len(a_list) == len(dates_list[0]) , 'page '+str(i+1)+' '+str(len(a_list))+' number of articles, '+str(len(dates_list[0]))+' number of dates'

    links_temp = [a['href'] for a in a_list]
    titles_temp = [a.text for a in a_list]
    sections_temp = dates_list[0]
    journalists_temp = [name.split(' ')[0] for name in dates_list[1]]
    dates_temp = dates_list[2]
    
    links += links_temp
    titles += titles_temp
    sections += sections_temp
    journalists += journalists_temp
    dates += dates_temp
    
    if i % 100 == 0:
        print(titles[-2:])
        print(sections[-2:])
        print(journalists[-2:])
        print(links[-2:])
    time.sleep(np.random.randint(50,300)/211)

df['title'] = titles
df['journalist'] = journalists
df['section'] = sections
df['link'] = links
df['date'] = dates
```

```python
for idx, link in df.iterrows():
    url = link['link']
    res_article = requests.get(url)
    res_article.encoding = 'utf-8'
    html_article_soup = BeautifulSoup(res_article.text, 'lxml')
    article_soup = html_article_soup.find('div',{'id':'article-view-content-div'})
    flag = True
    while flag:
        if len(article_soup('script')) > 0:
            article_soup('script')[0].extract()
        else:
            flag = False
    if len(article_soup.findAll('div',{'class':'view-copyright'})) > 0:
        article_soup.find('div',{'class':'view-copyright'}).extract()
    if len(article_soup.findAll('div',{'class':'view-editors'})) > 0:
        article_soup.find('div',{'class':'view-editors'}).extract()
    article = re.sub('\s\s+', ' ', re.sub('\n|\t|\r|\xa0', '', article_soup.text))
    df.loc[idx, 'article'] = article
    
    if idx % 100 == 0:
        print(df.loc[idx, 'title'])
    time.sleep(np.random.randint(50,200)/101)
```
