```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
import re
```

```python
df = pd.DataFrame(columns=['title','date','journalist','section','link','paper'])
links = []
titles = []
dates = []
journalists = []

for i in range(2526):
    links_temp = []
    titles_temp = []
    dates_temp = []
    journalists_temp = []
    
    url = 'https://biz.chosun.com/svc/list_in/list.html?catid=6&opt=t&pn='+str(i+1)
    res = requests.get(url)
    res.encoding='utf-8'
    html_soup = BeautifulSoup(res.text, 'lxml')
    
    a_list = html_soup.select('dt a')

    links_temp = [link['href'] for link in a_list]
    titles_temp = [link.text for link in a_list]
    dates_temp = [date.text for date in html_soup.select('span[class=date]')]
    journalists_temp = [journalist.text.split(' ')[0] for journalist in html_soup.select('span[class=author]')]
    
    links += links_temp
    titles += titles_temp
    dates += dates_temp
    journalists += journalists_temp
    
    if i % 100 == 0:
        print(titles_temp[0])
    time.sleep(np.random.randint(20,150)/101)

df['title'] = titles
df['link'] = links
df['date'] = dates
df['section'] = 'finance'
df['paper'] = 'chobiz'
df['journalist'] = journalists
df.to_csv('chobiz_finance_links.csv', encoding='utf-8', index=False)
```

```python
for idx, row in df.iterrows():
    url = 'http:' + row.link
    res = requests.get(url)
    res.encoding='utf-8'
    html_soup = BeautifulSoup(res.text, 'lxml')
    df.loc[idx, 'article'] = re.sub('\s\s+',' ',' '.join([text.text for text in html_soup.select('div[class=par]')]).strip())
    if idx % 100 == 0:
        print(row.title)
        if idx % 1000 == 0:
            df.to_csv('chobiz_finance_articles.csv', encoding='utf-8', index=False)
df.to_csv('chobiz_finance_articles.csv', encoding='utf-8', index=False)
```
